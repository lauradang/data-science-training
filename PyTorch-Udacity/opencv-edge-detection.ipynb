### Creating a custom kernel

The Sobel filter is very commonly used in edge detection and in finding patterns in intensity in an image. Applying a Sobel filter to an image is a way of **taking (an approximation) of the derivative of the image** in the x or y direction, separately.


## Kernel that Detects Horizontal Lines

```
sobel_y = np.array([
  [ -1, -2, -1], 
  [ 0, 0, 0], 
  [ 1, 2, 1]
])

filtered_image = cv2.filter2D(gray, -1, sobel_y)
plt.imshow(filtered_image, cmap='gray')
```

## Kernel that Detects Vertical Lines

```
sobel_x = np.array([
  [ -1, 0, 1], 
  [ -2, 0, 2], 
  [ -1, 0, 1]
])

filtered_image = cv2.filter2D(gray, -1, sobel_x)
plt.imshow(filtered_image, cmap='gray')
```

```
five_by_five = np.array([
  [ -2, -1, 0, 1, 2], 
  [ -3, -1.5, 0, 1.5, 3], 
  [ -4, -2, 0, 2, 4], 
  [ -3, -1.5, 0, 1.5, 3], 
  [ -2, -1, 0, 1, 2]
])

filtered_image = cv2.filter2D(gray, -1, five_by_five)
plt.imshow(filtered_image, cmap='gray')
```

## Blur Kernel

```
 # Multiplying each pixel by 1/9 then summing them up
 # That's the same as taking average of pixels
blur_kernel = np.array([
  [1/9, 1/9, 1/9],
  [1/9, 1/9, 1/9],
  [1/9, 1/9, 1/9]
])

blurred_image = cv2.filter2D(image, -1, blur_kernel)
plt.imshow(blurred_image)
```
